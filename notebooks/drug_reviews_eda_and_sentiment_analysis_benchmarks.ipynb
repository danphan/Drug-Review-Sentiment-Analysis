#### Let's do some minor preprocessing of the reviews for readability.

```python
data['review'] = data['review'].apply(lambda x: x[1:-1])
data['review'] = data['review'].apply(lambda x: x.replace('&#039;', "'"))
data['review'] = data['review'].apply(lambda x: x.replace('&amp;', "and"))
```

#### Let's restrict ourselves to the top condition, and use LSA and k-means to look at the various topics.

```python
condition = 'Birth Control'
data_bc = data[data['condition'] == condition]
data_bc.head()
```

#### Split data_bc into train and validation datasets

```python
plt.hist(data_bc['rating'],bins = np.arange(0.5,11.5));
plt.xlabel('Rating')
plt.ylabel('Number of reviews')
plt.title('{} Drug Review Ratings'.format(condition));
```

```python
STOP_WORDS = stopwords.words('english')
STOP_WORDS.remove('not') #'not' is a very important word for sentiment analysis!
condition_stop_words = ['birth','control','pill']
STOP_WORDS += ['ve','wa','ha']
# print(STOP_WORDS)

STOP_WORDS += condition_stop_words

#remove contractions from phrase
def decontracted(phrase):
    phrase = re.sub(r"\'t", "not", phrase)
    phrase = re.sub(r"\'re", " are", phrase)
    phrase = re.sub(r"\'s", " is", phrase)
    phrase = re.sub(r"\'d", " would", phrase)
    phrase = re.sub(r"\'ll", " will", phrase)
    phrase = re.sub(r"\'ve", " have", phrase)
    phrase = re.sub(r"\'m", " am", phrase)
    return phrase

lemmatizer = WordNetLemmatizer()

special_chars = '!#$%&()*+,-./:;<=>?@\\[]^_`{|}~\t\n\r'

# Remove special characters and stop words from review

def filter_review(review, stop=True):
    review = re.sub("[^a-zA-z\s']", '', review) #remove any special characters (keep apostrophes and spaces) 
    review = decontracted(review)
    # Change any white space to one space
    review = re.sub('\s+', ' ', review)
    review = review.split() #split on spaces
    review = [lemmatizer.lemmatize(word) for word in review] #Lemmatize reviews
    review = [word.lower() for word in review]
    
    if stop:
        review = [word for word in review if word not in STOP_WORDS]
        review = list(filter(lambda word: word not in STOP_WORDS, review))
    
    return ' '.join(review)
```

```python
data_bc['review_parsed'] = data_bc['review'].apply(filter_review)
```

#### Use TF-IDF to obtain most important words

```python
vectorizer = TfidfVectorizer(min_df = 5, max_df = 0.5,max_features=5000,ngram_range = (1,2))
t0 = time.time()
X_tfidf = vectorizer.fit_transform(data_bc['review_parsed'])

print(f"vectorization done in {time.time() - t0:.3f} s")
print(f"n_samples: {X_tfidf.shape[0]}, n_features: {X_tfidf.shape[1]}")
```

#### Sum the TF-IDF scores for each word over all of the reviews and sort from largest to smallest.

```python
importance = np.array(X_tfidf.sum(axis=0))[0]
#obtain the indices which sort the words from most to least important
idx_sort = np.argsort(importance)[::-1]
words_sorted = np.array(vectorizer.get_feature_names())[idx_sort]
word_score_sorted = importance[idx_sort]
#print(words_sorted[:200])

num_words =25
plt.plot(words_sorted[:num_words],word_score_sorted[:num_words],'o')
plt.xlabel('Word')
plt.ylabel('Word Importance')
plt.xticks(rotation=45,ha='right');
```

### Apply Latent Semantic Analysis (LSA), then K-means clustering to see what topics are being discussed in the reviews

```python
#apply SVD for dimensionality reduction, then normalize prior to k-means clustering
lsa = make_pipeline(TruncatedSVD(n_components=100), Normalizer(copy=False))
t0 = time.time()
X_lsa = lsa.fit_transform(X_tfidf)
explained_variance = lsa[0].explained_variance_ratio_.sum()

print(f"LSA done in {time.time() - t0:.3f} s")
print(f"Explained variance of the SVD step: {explained_variance * 100:.1f}%")
```

```python
Sum_of_squared_distances = []
K = range(1,20)
for num_clusters in K :
    kmeans = KMeans(n_clusters=num_clusters,max_iter = 100, n_init = 20)
    kmeans.fit(X_lsa)
    Sum_of_squared_distances.append(kmeans.inertia_)
plt.plot(K,Sum_of_squared_distances,'bx-')
plt.xlabel('Values of K') 
plt.ylabel('Sum of squared distances/Inertia') 
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
range_n_clusters = range(2,20)
silhouette_avg = []
for num_clusters in range_n_clusters:
 
    # initialise kmeans
    kmeans = KMeans(n_clusters=num_clusters,max_iter = 100, n_init = 20)
    kmeans.fit(X_lsa)
    cluster_labels = kmeans.labels_
 
    # silhouette score
    silhouette_avg.append(silhouette_score(X_lsa, cluster_labels))
plt.plot(range_n_clusters,silhouette_avg,"bx-")
plt.xlabel("Values of K")
plt.ylabel("Silhouette score") 
plt.title('Silhouette analysis For Optimal k')
plt.show()
```

Take k = 5, perform k-means clustering, and look at the top 10 words in each cluster.

```python
# initialise kmeans
true_k = 5
kmeans = KMeans(n_clusters=true_k, n_init = 20, max_iter=100)
kmeans.fit(X_lsa)
cluster_labels = kmeans.labels_

#obtain the centers of the cluster in latent space
#then invert the SVD/LSA to find what kind of documents these clusters correspond to 
original_space_centroids = lsa[0].inverse_transform(kmeans.cluster_centers_) #cluster-word matrix
order_centroids = original_space_centroids.argsort()[:, ::-1] #sort the words in each cluster by importance
terms = vectorizer.get_feature_names()

for i in range(true_k):
    print(f"Cluster {i+1}: ", end="")
    for ind in order_centroids[i, :10]:
        print(f"{terms[ind]}, ", end="")
    print()
```

#### See how the ratings are distributed in each cluster

```python
data_bc['cluster_labels'] = kmeans.predict(X_lsa)

fig, ax = plt.subplots(nrows = 1, ncols = 5, figsize = (12,5),sharex=True,sharey=True)

for i in range(true_k):
    data_cluster = data_bc[data_bc['cluster_labels'] == i]
    ax[i].hist(data_cluster['rating'],
             bins =np.arange(0.5,11.5),\
             histtype='step',
             density = True)
    ax[i].set_title('Cluster {}'.format(i+1))

    ax[i].set_xticks(range(1,11));
```

### Apply Latent Dirichlet Allocation (LDA) for topic modeling

```python
# Use tf (raw term count) features for LDA.
tf_vectorizer = CountVectorizer(
    max_df=0.7, min_df=5, max_features=5000
)
tf = tf_vectorizer.fit_transform(data_bc['review_parsed'])
```

```python
n_topics = 3

lda = LatentDirichletAllocation(
    n_components=n_topics,
    max_iter=5,
    learning_method="online",
    learning_offset=50.0,
    random_state=0,
)
lda.fit(tf)
```

```python
tf_feature_names = tf_vectorizer.get_feature_names()
n_top_words = 20

def plot_top_words(model, feature_names, n_top_words, title):
    fig, axes = plt.subplots(1, n_topics, figsize=(30, 15), sharex=True)
    axes = axes.flatten()
    for topic_idx, topic in enumerate(model.components_):
        top_features_ind = topic.argsort()[: -n_top_words - 1 : -1]
        top_features = [feature_names[i] for i in top_features_ind]
        weights = topic[top_features_ind]

        ax = axes[topic_idx]
        ax.barh(top_features, weights, height=0.7)
        ax.set_title(f"Topic {topic_idx +1}", fontdict={"fontsize": 30})
        ax.invert_yaxis()
        ax.tick_params(axis="both", which="major", labelsize=20)
        for i in "top right left".split():
            ax.spines[i].set_visible(False)
        fig.suptitle(title, fontsize=40)

    plt.subplots_adjust(top=0.90, bottom=0.05, wspace=0.90, hspace=0.3)
    plt.show()
    
plot_top_words(lda, tf_feature_names, n_top_words, "Topics in LDA model")
```

### Define function which returns the average accuracy and f1 scores over the training set, using k-fold cross-validation

```python
def cross_validate_model(model, X, y, n_splits = 10):
    cv = StratifiedKFold(n_splits = n_splits)

    acc_scores = []
    f1_scores = []

    for fold, (train, val) in enumerate(cv.split(X,y)):
        print('Fold {}:'.format(fold))
        model.fit(X[train],y.iloc[train])
        y_pred = model.predict(X[val])
        y_val = y.iloc[val]
        acc = accuracy_score(y_pred, y_val)
        f1 = f1_score(y_pred,y_val)

        acc_scores.append(acc)
        f1_scores.append(f1)
        
        print('Accuracy: {:.2f}%'.format(100*np.mean(acc_scores)))
        print('F1 score: {:.2f}%'.format(100*np.mean(f1_scores)))

    acc_scores = np.array(acc_scores)
    f1_scores = np.array(f1_scores)
    print('\nAccuracy: {:.2f}+-{:.2f}%'.format(100*np.mean(acc_scores),100*np.std(acc_scores)))
    print('F1 score: {:.2f}+-{:.2f}%'.format(100*np.mean(f1_scores),100*np.std(f1_scores)))
```

### Logistic Regression (using the compressed LSA representation of each document as our features.)

```python
#scale features before inputting to Logistic Regression. Important when features are regularized.

model_logreg = make_pipeline(StandardScaler(),LogisticRegression())

cross_validate_model(model_logreg, X_lsa, data_bc['sentiment'])
```

### Naive Bayes (using the TF-IDF document-term matrix as input). 

```python
model_nb = MultinomialNB()

cross_validate_model(model_nb, X_tfidf, data_bc['sentiment'])
```

### SVMs

```python
model_svm = make_pipeline(StandardScaler(),LinearSVC(C = 1e-1))

cross_validate_model(model_logreg, X_lsa, data_bc['sentiment'],n_splits = 10)
```

### Random Forests

```python
model_rf = RandomForestClassifier()
cross_validate_model(model_rf, X_tfidf, data_bc['sentiment'])
```
